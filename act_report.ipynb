The goal of this study was to investigate data from the WeRateDogs Twitter account. The study aims to answer questions such as "Which name for a dog is the most popular among competitors, what stage has the highest value, and what stage favourite countÂ has the highest mean?"

The investigation is centred on three primary components:
1. Names column from WeRateDogs. The top ten (10) names were chosen from a large number of dogs.
2. The second set of information comes from the dog stages table.
3. The third part is based on each dog stage's favourite count.

### Analysis

#### Insight 1: What is the name of the most common dog in the data


```python
import pandas as pd
import matplotlib
df = pd.read_csv('twitter_archive_master.csv')
```

```python
df['name'].value_counts().nlargest(10).plot(kind = 'bar');
df['name'].value_counts().nlargest(10)
```

The most common dog name in this data is Charlie with value count of 11. Oliver, Lucy, and Cooper came second with count of 10 each, while Penny and Tucker came Third. These are the names among the top ten names.

#### Insight 2: What stage has the hightest number of dogs that partake in the tweet

```python
df['dog_stages'].value_counts().plot(kind = 'pie',figsize=(8,8), autopct='%.2f%%')
df['dog_stages'].value_counts()
```

The dog state with the highest value counts is Pupper with 66.10% (117) of all the dogs that partake in the data. Doggo came second with 20.34% (36) while Puppo came third with 8.47% (15)

#### Insight 3: What stage favourite count has the highest mean?

```python
df.groupby('dog_stages')['favorite_count'].mean().plot(kind='barh')

df.groupby('dog_stages')['favorite_count'].mean()
```

Doggo has the highest mean of favorite count. Despite Pupper with the highest value count, The mean Pupper mean of favorite count fall behind that of doggo, but ahead of other stages.
